```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

import sys  
sys.path.insert(1, '../python_files/')
import pulses
```

```python
def f(A): return A/30
f_supp = 2

testpulse = pulses.Pulse(f,f_supp,[pulses.zero,pulses.linear(5,0,0),pulses.linear_opp(5,0,125),pulses.zero],[10000,25000,25000,10000],time_unit=0.001, drag=True)
```

```python
fig1=plt.figure()
plt.plot(testpulse.time_vals,testpulse.raw)
plt.xlabel("Aika (ms)")
plt.ylabel("Normalisoitu amplitudi")
plt.title("Alkuperäinen pulssi")
plt.show()

freqs = np.fft.rfftfreq(len(testpulse.time_vals), d=0.001)
fourier = np.fft.rfft(testpulse.raw)

fig2=plt.figure()
plt.semilogy(freqs*2*np.pi, np.abs(fourier)/np.max(np.abs(fourier)))
plt.xlim([0,20])
plt.xlabel("Taajuus (KHz)")
plt.ylabel("Normalisoitu amplitudi")
plt.title("Fouriermuunnos alkuperäisestä pulssista")
plt.show()

fig4=plt.figure()
plt.plot(testpulse.time_vals,testpulse.drag_pulse)
plt.plot(testpulse.time_vals,testpulse.drag_component)
plt.legend(["Pulssi","DRAG-komponentti"])
plt.xlabel("Aika (ms)")
plt.ylabel("Normalisoitu amplitudi")
plt.title("DRAG-korjattu pulssi")
plt.show()

drag_fourier = np.fft.rfft(testpulse.drag_pulse)

fig5=plt.figure()
plt.semilogy(freqs*2*np.pi, np.abs(drag_fourier)/np.max(np.abs(drag_fourier)))
plt.xlim([0,20])
plt.xlabel("Taajuus (KHz)")
plt.ylabel("Normalisoitu amplitudi")
plt.title("Fouriermuunnos DRAG-korjatusta pulssista")
plt.show()
```
